```python
import pandas
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
fname = "csv_data/dirty_data/spy_dailys.csv"
df = pandas.read_csv(fname)
```

```python
# add return column and a time column
df['Return'] = df['Close'] - df['Open']
df['Unix Timestamp'] = [datetime.strptime(elt, '%Y-%m-%d').timestamp() for elt in df['Date']]
```

```python
df.head()
```

```python
timestamp = np.asarray(df["Unix Timestamp"])
spy_open = np.asarray(df["Open"])
spy_close = np.asarray(df["Close"])
```

```python
plt.plot(timestamp, spy_open, label="SPY open")
plt.plot(timestamp, spy_close, label="SPY close")

# plt.plot(timestamp, returns, label='daily returns')

plt.xlabel('Unix timestamp')
plt.ylabel("price")
plt.title("SPY 1 yr plot")
plt.legend()
plt.show()
```

```python
returns = np.asarray(df["Return"])

plt.plot(timestamp, returns, label='daily returns')
plt.xlabel('Unix timestamp')
plt.ylabel("price")
plt.title("SPY 1 yr plot")
plt.legend()
plt.show()
```
